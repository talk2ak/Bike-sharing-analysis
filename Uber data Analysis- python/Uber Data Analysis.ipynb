```python
# convert date into day, year, month

uber_df.insert(loc=2,column="Year",value=uber_df['START_DATE*'].dt.year)

uber_df.insert(loc=3,column="Month",value=uber_df['START_DATE*'].dt.month)

uber_df.insert(loc=4,column="Date",value=uber_df['START_DATE*'].dt.date)

uber_df.insert(loc=5,column="Day",value=uber_df['START_DATE*'].dt.dayofweek+1)

uber_df.insert(loc=6,column="Hour",value=uber_df['START_DATE*'].dt.hour)

```

```python
# Outlier detection
# 1. outlier should be removed with out loosing too much data

import seaborn as sns
sns.boxplot(uber_df['MILES*'])

# this is visualization by which we can calculate outlier
```

```python
df=uber_df.copy()
```

```python
IQR=df['MILES*'].quantile(0.75)-df['MILES*'].quantile(0.25)
```

```python
IQR
```

```python
lower=df['MILES*'].quantile(0.25)-(IQR*1.5)
upper=df['MILES*'].quantile(0.75)+(IQR*1.5)
print(lower)
print(upper)
```

```python
df[df['MILES*']>21.65].count()
```

```python
df[df['MILES*']<-8.35].count()
```

```python
df[df['MILES*']>21.65] #outlier data
```

```python
# Conditional Data deletion

df.drop(df[df['MILES*']>21.65].index,inplace=True)
```
